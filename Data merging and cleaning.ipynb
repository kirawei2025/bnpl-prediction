```python
# Please add any packages here and freeze any to the requirements.txt file
import numpy as np
import pandas as pd

# Datasets as separate years
data_21 = pd.read_csv("public2021.csv")
data_22 = pd.read_csv("public2022.csv")
data_23 = pd.read_csv("public2023.csv")
data_24 = pd.read_csv("public2024.csv")
```

The wording and order of the BNPL variables changed between 2021 and 2022-onwards so I reordered the 2021 dataset to merge it properly with the other years.

```python
data_21["BNPL4_d"] = data_21["BNPL4_b"]
data_21["BNPL4_e"] = data_21["BNPL4_c"]
```

```python
datasets = [data_21, data_22, data_23, data_24]
datasets_transformed = []
years = [2021,2022,2023,2024]

# Selects relevant columns and combines datasets, organized by year
for i, dataset in enumerate(datasets):
    dataset = dataset[["BNPL1","BNPL3","BNPL4_a","BNPL4_b","BNPL4_c","BNPL4_d","BNPL4_e","BNPL4_f",
            "EF3_a","EF3_b","EF3_c","EF3_d","EF3_e","EF3_f","EF3_g","EF3_h","I20",
            "ppfs0596",
            "ppeducat","ppage","ppagecat","ppethm","ppgender","ppinc7","ppemploy","ppmarit5",
            "SL6",
            "B3", "A6", "C4A", "ppage", "EF1", "ppfs1482"]]
    dataset["year"] = years[i]
    datasets_transformed.append(dataset)
data_merged = pd.concat(datasets_transformed)
```
