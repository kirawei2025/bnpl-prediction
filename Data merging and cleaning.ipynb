## Merge data sets

```python
# Please add any packages here and freeze any to the requirements.txt file
import numpy as np
import pandas as pd

# Datasets as separate years
data_21 = pd.read_csv("public2021.csv")
data_22 = pd.read_csv("public2022.csv")
data_23 = pd.read_csv("public2023.csv")
data_24 = pd.read_csv("public2024.csv")
```

The wording and order of the BNPL variables changed between 2021 and 2022-onwards so I reordered the 2021 dataset to merge it properly with the other years.

```python
# Rearranging the columns in the 2021 data to match the current survey format
data_21["BNPL4_f"] = data_21["BNPL4_d"]
data_21["BNPL4_d"] = data_21["BNPL4_b"]
data_21["BNPL4_e"] = data_21["BNPL4_c"]

# Setting columns with missing data to nan. Can undo this later if we would like the 2021 specific variables.
data_21["BNPL4_a"] = np.nan
data_21["BNPL4_b"] = np.nan
data_21["BNPL4_c"] = np.nan
```

```python
# create columns X12_abcdefg for 21-23 datasets filling with na
# X12: financial concerns
data_21[["X12_a", "X12_b", "X12_c", "X12_d", "X12_e", "X12_f", "X12_g"]] = np.nan
data_22[["X12_a", "X12_b", "X12_c", "X12_d", "X12_e", "X12_f", "X12_g"]] = np.nan
data_23[["X12_a", "X12_b", "X12_c", "X12_d", "X12_e", "X12_f", "X12_g"]] = np.nan
```

```python
# create columns EF5C for 21-22 datasets filling with na
# EF5C: did you pay all your bills in full last month?
data_21[["EF5C"]] = np.nan
data_22[["EF5C"]] = np.nan
```

```python
# LOC is a mix of str and float answers. covert all to str.
data_21['L0C'] = data_21['L0C'].apply(lambda x: str(int(x)) if isinstance(x,float) and not pd.isna(x) else x)
data_22['L0C'] = data_22['L0C'].apply(lambda x: str(int(x)) if isinstance(x,float) and not pd.isna(x) else x)
data_23['L0C'] = data_23['L0C'].apply(lambda x: str(int(x)) if isinstance(x,float) and not pd.isna(x) else x)
data_24['L0C'] = data_24['L0C'].apply(lambda x: str(int(x)) if isinstance(x,float) and not pd.isna(x) else x)
```

```python
# convert 5,6,7, 7 or more in 2021 and 2022 data to 5 or more to sync with 2023, 2024 data
data_21['L0C'] = data_21['L0C'].apply(lambda x: '5 or more' if x in ['5','6','7','7 or more'] else x)
data_22['L0C'] = data_22['L0C'].apply(lambda x: '5 or more' if x in ['5','6','7','7 or more'] else x)
```

```python
# convert EF3_abcdefgh values to be consistent across years
mapping = {
  'Put it on my credit card and pay it off in full at the next statement': 'Yes',
  'Put it on my credit card and pay it off over time': 'Yes',
  'With the money currently in my checking/savings account or with cash': 'Yes',
  'Using money from a bank loan or line of credit': 'Yes',
  'By borrowing from a friend or family member': 'Yes',
  'Using a payday loan, deposit advance, or overdraft':'Yes',
  'By selling something': 'Yes',
  "I wouldn’t be able to pay for the expense right now": 'Yes',
  'No, did not do this': 'No'
}

cols = ["EF3_a", "EF3_b", "EF3_c", "EF3_d", "EF3_e", "EF3_f", "EF3_g", "EF3_h"]
data_21[cols] = data_21[cols].apply(lambda col: col.map(mapping))
data_22[cols] = data_22[cols].apply(lambda col: col.map(mapping))
data_23[cols] = data_23[cols].apply(lambda col: col.map(mapping))
data_24[cols] = data_24[cols].apply(lambda col: col.map(mapping))
```

```python
# Merging and exporting the dataset
datasets = [data_21, data_22, data_23, data_24]
datasets_transformed = []
years = [2021,2022,2023,2024]

# Selects relevant columns and combines datasets, organized by year
for i, dataset in enumerate(datasets):
    dataset = dataset[["weight","BNPL1","BNPL3","BNPL4_a","BNPL4_b","BNPL4_c","BNPL4_d","BNPL4_e","BNPL4_f",
            "EF3_a","EF3_b","EF3_c","EF3_d","EF3_e","EF3_f","EF3_g","EF3_h","I20",
            "ppfs0596","ppeducat","ppage","ppagecat","ppethm","ppgender","ppinc7","ppemploy","ppmarit5","pphhsize","ppkid017", "ppfs1482",
            "SL6","B3","A6","C4A", "EF1","L0C","I41_c","I41_e",'A1_a','A1_b','A1_c','X12_a','X12_b','X12_c','X12_d','X12_e','X12_f','X12_g',
            'EF5C']]
    dataset["year"] = years[i]
    datasets_transformed.append(dataset)
data_merged = pd.concat(datasets_transformed)
```

```python
# Export
display(data_merged)
data_merged.to_csv("data/data_merged.csv")
```
