## Merge data sets

### Import datasets

```python
# Please add any packages here and freeze any to the requirements.txt file
import numpy as np
import pandas as pd

# Datasets as separate years
data_22 = pd.read_csv("../../data/public2022.csv")
data_23 = pd.read_csv("../../data/public2023.csv")
data_24 = pd.read_csv("../../data/public2024.csv")
data_25 = pd.read_csv("../../data/public2025.csv")
```

### Data pre-processing

```python
# change 2025 ppfsasset to ppfs0596 to sync with other years
[df.rename(columns = {'ppfs0596': 'ppfsasset'}, inplace=True) for df in [data_22,data_23,data_24]]
```

```python
# list of predictive columns to use
cols = ["weight","BNPL1","BNPL3","BNPL4_a","BNPL4_b","BNPL4_c","BNPL4_d","BNPL4_e","BNPL4_f",
            "EF3_a","EF3_b","EF3_c","EF3_d","EF3_e","EF3_f","EF3_g","EF3_h","I20",
            "ppfsasset","ppeducat","ppage","ppethm","ppgender","ppinc7","ppemploy","ppmarit5", "ppfs1482",
            "A6","C4A", "EF1","L0C","I41_c","I41_e",'A1_a','A1_b','A1_c','X12_a','X12_b','X12_c','X12_d','X12_e','X12_f','X12_g',
            'EF5C','C3P','K21_a','INF4',
            'BNPL1A','BNPL5','EF5D']
```

```python
# check all columns in each dataset
cols_2022= data_22.columns
cols_2023= data_23.columns
cols_2024= data_24.columns
cols_2025= data_25.columns
```

```python
# check existance of predictive columns in each dataset
missing_cols_2022 = list(set(cols) - set(cols_2022))
missing_cols_2023 = list(set(cols) - set(cols_2023))
missing_cols_2024 = list(set(cols) - set(cols_2024))
missing_cols_2025 = list(set(cols) - set(cols_2025))
# add missing columns filled with na
data_22[missing_cols_2022] = np.nan
data_23[missing_cols_2023] = np.nan
data_24[missing_cols_2024] = np.nan
data_25[missing_cols_2025] = np.nan
```

```python
# combine 4 datasets
data = []
datasets = [data_22, data_23, data_24, data_25]
years = [2022,2023,2024,2025]
for i, d in enumerate(datasets):
  ds = d[cols]
  ds['year'] = years[i]
  data.append(ds)
data_merged = pd.concat(data)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# check values in each column across all years
data_check = []
for c in cols:
  data_check.append({
    'column': c,
    'data_2025': data_merged[data_merged['year'] == 2025][c].unique(),
    'data_2024': data_merged[data_merged['year'] == 2024][c].unique(),
    'data_2023': data_merged[data_merged['year'] == 2023][c].unique(),
    'data_2022': data_merged[data_merged['year'] == 2022][c].unique()
  })
pd.DataFrame(data_check)
```

```python
# modify the values to sychronize across all datasets
  # BNPL1 in 2025 "Most of time"
mask_year_2025 = data_merged['year']==2025
data_merged.loc[mask_year_2025, 'BNPL1'] = data_merged.loc[mask_year_2025, 'BNPL1'].apply(lambda x: 'Yes' if x=='Most or all of the time' else x)
  # EF3 in 2022-2023 "Yes"
mask_year = data_merged['year'].isin([2022,2023])
mask_cols = ['EF3_a','EF3_b','EF3_c','EF3_d','EF3_e','EF3_f','EF3_g','EF3_h']
data_merged.loc[mask_year, mask_cols] = data_merged.loc[mask_year, mask_cols].map(lambda x: 'No' if x== 'No' else 'No' if x =='No, did not do this' else 'Yes')
  # ppgender in 2025 remove non gender answers
data_merged = data_merged[data_merged['ppgender'].isin(['Male','Female'])]
  # ppfsasset in 2025 - sync with other years
mapping_ppfsasset = {
  'Under $25,000': 'Under $50,000',
  '$25,000 - $49,999': 'Under $50,000',
  '$1,000,000 - $1,999,999': '$1,000,000 or more',
  '$2,000,000 or more': '$1,000,000 or more'
}
mask_year_2025 = data_merged['year']==2025
data_merged.loc[mask_year_2025, 'ppfsasset'] = data_merged.loc[mask_year_2025, 'ppfsasset'].replace(mapping_ppfsasset)
  # L0C - convert all to str
mask_year_2224 = data_merged['year'].isin([2022,2024])
data_merged.loc[mask_year_2224, 'L0C'] = data_merged.loc[mask_year_2224, 'L0C'].apply(lambda x: int(x) if isinstance(x, float) and not pd.isna(x) else x).astype(str)
mask_year_2022 = data_merged['year'].isin([2022])
data_merged.loc[mask_year_2022, 'L0C'] = data_merged.loc[mask_year_2022, 'L0C'].replace({
  '5':'5 or more',
  '6': '5 or more',
  '7 or more': '5 or more'
}, inplace=True
)
  # C3P 2023,2024 - capitalize
mask_year_2324= data_merged['year'].isin([2024,2023])
data_merged.loc[mask_year_2324, 'C3P'] = data_merged.loc[mask_year_2324, 'C3P'].str.capitalize()
```
